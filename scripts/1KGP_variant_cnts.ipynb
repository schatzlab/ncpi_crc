```python
# Import necessary libraries
import vcf
import pandas as pd
import matplotlib.pyplot as plt
import re

# Function to parse the VCF file and count variants per chromosome
def count_variants_per_chromosome(vcf_file):
    # Create a dictionary to store the variant counts per chromosome
    chromosome_variants = {}
    
    # Open the VCF file
    vcf_reader = vcf.Reader(open(vcf_file, 'r'))
    
    # Iterate through each record in the VCF file
    for record in vcf_reader:
        # Get the chromosome
        chrom = record.CHROM
        
        # Increment the variant count for this chromosome
        if chrom in chromosome_variants:
            chromosome_variants[chrom] += 1
        else:
            chromosome_variants[chrom] = 1
    
    return chromosome_variants

# Custom sorting function to handle chromosomes in numerical order
def chromosome_sort_key(chrom):
    # Extract numbers from chromosomes, handling "X", "Y", "MT" separately
    if chrom == "X":
        return 23
    elif chrom == "Y":
        return 24
    elif chrom == "MT" or chrom == "M":
        return 25
    else:
        # Use regular expressions to extract numerical values
        match = re.match(r'(\d+)', chrom)
        if match:
            return int(match.group(1))
        else:
            return 26  # Handle any unexpected chromosome names

# Function to plot the variant counts per chromosome
def plot_variants_per_chromosome(variant_counts):
    # Convert the variant counts dictionary to a DataFrame for easy plotting
    df = pd.DataFrame(list(variant_counts.items()), columns=['Chromosome', 'Variant_Count'])
    
    # Sort the chromosomes using the custom sorting function
    df['Chromosome'] = df['Chromosome'].astype(str)
    df = df.sort_values(by='Chromosome', key=lambda x: x.map(chromosome_sort_key))
    
    # Plot the data
    plt.figure(figsize=(10,6))
    plt.bar(df['Chromosome'], df['Variant_Count'], color='skyblue')
    plt.xlabel('Chromosome')
    plt.ylabel('Number of Variants')
    plt.title('Number of Variants per Chromosome')
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()

# Example usage
vcf_file = '1000G_omni2.5.hg38.vcf'
variant_counts = count_variants_per_chromosome(vcf_file)
plot_variants_per_chromosome(variant_counts)
```
